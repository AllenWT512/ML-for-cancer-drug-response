```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Encoding the target variable
label_encoder = LabelEncoder()
Drug1_Gene_MLdataset['Drug1_response'] = label_encoder.fit_transform(Drug1_Gene_MLdataset['Drug1_response'])

# Separating features and target variable
X_1 = Drug1_Gene_MLdataset.drop(['Drug1_response'], axis=1)
y_1 = Drug1_Gene_MLdataset['Drug1_response']

# Splitting the dataset into training and testing sets
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2, random_state=42)

# Applying SMOTE to deal with class imbalance
smote = SMOTE(random_state=42)
X_train_1_balanced, y_train_1_balanced = smote.fit_resample(X_train_1, y_train_1)
```

```python
Drug1_Gene_MLdataset['Drug1_response'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report

# Initialize the Logistic Regression classifier with class weight balanced
log_reg = LogisticRegression(class_weight='balanced', max_iter=100, random_state=42)

# Train the classifier
log_reg.fit(X_train_1_balanced, y_train_1_balanced)

# Perform cross-validation and calculate accuracy
cv_scores = cross_val_score(log_reg, X_train_1_balanced, y_train_1_balanced, cv=10, scoring='accuracy')

# Predict on the test set
y_pred_1 = log_reg.predict(X_test_1)
cv_scores.mean()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

precision_1 = precision_score(y_test_1, y_pred_1)
recall_1 = recall_score(y_test_1, y_pred_1)
f1_score_1 = f1_score(y_test_1, y_pred_1)
auc_1 = roc_auc_score(y_test_1, y_pred_1)
test_accuracy_1 = accuracy_score(y_test_1, y_pred_1)

print("Drug1 response prediction results:")
print("Precision is", precision_1)
print("recall is", recall_1)
print("f1_score is", f1_score_1)
print("AUC_score is", auc_1)
print("Accuracy is", test_accuracy_1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import classification_report

# Initialize the Logistic Regression classifier with class weight balanced
log_reg = LogisticRegression(penalty="elasticnet", solver='saga', class_weight='balanced', l1_ratio=0.5, max_iter=100, random_state=42)

# Train the classifier
log_reg.fit(X_train_4_balanced, y_train_4_balanced)

# Perform cross-validation and calculate accuracy
cv_scores = cross_val_score(log_reg, X_train_4_balanced, y_train_4_balanced, cv=10, scoring='accuracy')

# Predict on the test set
y_pred_4 = log_reg.predict(X_test_4)

cv_scores.mean()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

precision_4 = precision_score(y_test_4, y_pred_4)
recall_4 = recall_score(y_test_4, y_pred_4)
f1_score_4 = f1_score(y_test_4, y_pred_4)
auc_4 = roc_auc_score(y_test_4, y_pred_4)
test_accuracy_4 = accuracy_score(y_test_4, y_pred_4)
print("Drug4 response prediction results:")
print("Precision is", precision_4)
print("recall is", recall_4)
print("f1_score is", f1_score_4)
print("AUC_score is", auc_4)
print("Accuracy is", test_accuracy_4)
```
