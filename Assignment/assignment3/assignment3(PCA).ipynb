**3.1 Predict the drug response of all four drugs based on the mutational and copy number data**

```python
import pandas as pd # 
import numpy as np
Drug1_Gene_dataset = pd.read_csv("Drug1_Gene_dataset.csv", index_col = 0)
Drug2_Gene_dataset = pd.read_csv("Drug2_Gene_dataset.csv", index_col = 0)
Drug3_Gene_dataset = pd.read_csv("Drug3_Gene_dataset.csv", index_col = 0)
Drug4_Gene_dataset = pd.read_csv("Drug4_Gene_dataset.csv", index_col = 0)
```

```python
Drug1_Gene_dataset.loc[:,"Drug1_response"] = Drug1_Gene_dataset["Drug1_response"].apply(lambda x: 0 if x=="Sensitivity" else 1)
Drug2_Gene_dataset.loc[:,"Drug2_response"] = Drug2_Gene_dataset["Drug2_response"].apply(lambda x: 0 if x=="Sensitivity" else 1)
Drug3_Gene_dataset.loc[:,"Drug3_response"] = Drug3_Gene_dataset["Drug3_response"].apply(lambda x: 0 if x=="Sensitivity" else 1)
Drug4_Gene_dataset.loc[:,"Drug4_response"] = Drug4_Gene_dataset["Drug4_response"].apply(lambda x: 0 if x=="Sensitivity" else 1)
```

```python
Drug1_Gene_dataset.reset_index(inplace = True,drop = True)
Drug2_Gene_dataset.reset_index(inplace = True,drop = True)
Drug3_Gene_dataset.reset_index(inplace = True,drop = True)
Drug4_Gene_dataset.reset_index(inplace = True,drop = True)
```

```python
Drug1_Gene_dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Encoding the target variable
label_encoder = LabelEncoder()
Drug1_Gene_dataset['Drug1_response'] = label_encoder.fit_transform(Drug1_Gene_dataset['Drug1_response'])

# Separating features and target variable
X_1 = Drug1_Gene_dataset.drop(['Drug1_response'], axis=1)
y_1 = Drug1_Gene_dataset['Drug1_response']
X_1.shape
```

```python
Drug1_Gene_dataset
```

```python
pca_1 = PCA(n_components=0.95,random_state = 40)  
X_1_pca = pca_1.fit_transform(X_1)
X_1_pca.shape, pca_1.n_components_
```

```python
print(pca_1.explained_variance_ratio_)
```

```python
loadings_1 = pca_1.components_.T
loadings_1
```

```python
features_1 = X_1.columns
loadings_df_1 = pd.DataFrame(loadings_1, index=features_1)
loadings_df_1
```

```python
loadings_df_1 = pd.DataFrame(loadings_1, columns=[f'PC{i+1}' for i in range(pca_1.n_components_)], index=features_1)
loadings_df_1
```

**What are the important features?**

```python
important_features_list_1 = []
for i, pc in enumerate(loadings_df_1.columns):
    idx_max = np.argmax(np.abs(loadings_df_1[pc]))
    important_features_list_1.append({
        'PC': pc,
        'Drug1 Important Feature': loadings_df_1.index[idx_max],
        'Loading': loadings_df_1[pc][idx_max]
    })
important_features_df_1 = pd.DataFrame(important_features_list_1)
print(important_features_df_1)
```

```python
# Splitting the dataset into training and testing sets
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1_pca, y_1, test_size=0.2, stratify=y_1, random_state=42)
y_train_1.value_counts()
```

```python
# Applying SMOTE to deal with class imbalance
smote = SMOTE(random_state=42)
X_train_1_balanced, y_train_1_balanced = smote.fit_resample(X_train_1, y_train_1)
```

```python
import warnings
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import classification_report
warnings.filterwarnings('ignore')


log_reg = LogisticRegression(class_weight='balanced', random_state=41)  # Initialize the Logistic Regression classifier with class weight balanced
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], # Regularization strength
    'penalty': ['l1', 'l2', 'elasticnet'],  # Type of regularization, the function is to reduce the overfitting
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],   # Algorithm for optimization, to find the optimal coefficients based on minimum loss function
    'max_iter': [100, 200, 300, 400, 500],  # Maximum number of iterations taken for the solvers to converge
}
grid_search = GridSearchCV(log_reg, param_grid, cv=10, scoring='accuracy', verbose = 0) # Initialize the grid search with 10 fold cross-validation and accuracy as the scoring metric, verbose means to print the progress

grid_search.fit(X_train_1_balanced, y_train_1_balanced) # Fit the model , Train the classifier

# Perform cross-validation and calculate accuracy
#cv_scores = cross_val_score(log_reg, X_train_1_balanced, y_train_1_balanced, cv=10, scoring='accuracy')

# Predict on the test set
y_pred_1 = grid_search.predict(X_test_1)
best_score = grid_search.best_score_
best_score
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

precision_1 = precision_score(y_test_1, y_pred_1)
recall_1 = recall_score(y_test_1, y_pred_1)
f1_score_1 = f1_score(y_test_1, y_pred_1)
auc_1 = roc_auc_score(y_test_1, y_pred_1)
test_accuracy_1 = accuracy_score(y_test_1, y_pred_1)
print("Drug1 response prediction results:")
print("Precision is", precision_1)
print("recall is", recall_1)
print("f1_score is", f1_score_1)
print("AUC_score is", auc_1)
print("Accuracy is", test_accuracy_1)
```
